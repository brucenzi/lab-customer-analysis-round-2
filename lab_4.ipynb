```python
#as there was a significant amount of NaNs in each of the previous columns, I decided to assume they were "not informed" rather than redistribute it to other possible values. 
df = df.fillna(0)
df.isna().sum()
```

```python
#6 Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.
```

```python
#create a new column extracting the month only from the column "effective to date", which is the only one displaying dates in the dataframe
df['effective to date (months)'] = pd.DatetimeIndex(df['effective to date']).month
display(df.head())
print('New column "efective to date (months)" data type is:')
print(df['effective to date (months)'].dtype)
```

```python
#to filter the Q1 months, I used the operation <4, as they are now set as numerical 
display(df.loc[(df['effective to date (months)']<4)])
print("Filtered column with total Q1 values")
print(df["effective to date (months)"].value_counts())
```
